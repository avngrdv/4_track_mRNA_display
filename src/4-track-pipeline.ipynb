This notebook reproduces a deep learning pipeline to train substrate preference models for F<sub>420</sub>H<sub>2</sub>-dependent dehydroamino acid reductases from 4-track mRNA display NGS data (`DDBJ #DRA018924`).

Everywhere in the code, npAA abbreviations are as follows: `d`= Dha; `a`= N<sup>Me</sup>Ala

To reproduce the code below, merged R1/R2 NGS files can be added to `../sequencing_data` to start. The files are too big and are not stored in this repo.

```python
#import pre-requisites and set up the pipeline

import os
os.environ['TF_XLA_FLAGS'] = '--tf_xla_enable_xla_devices'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '1' 

import numpy as np
np.set_printoptions(suppress=True, precision=3)

from clibas.parsers import FastqParser
from clibas.pipelines import Pipeline
from clibas.dataanalysis import DataAnalysisTools
from clibas.dispatchers import Dispatcher
from clibas.datapreprocessors import DataPreprocessor
from clibas.classifier import Classifier

#config file holds the information about library designs and
#other parser instructions (where to look for data, where to save results etc)
import ADH_config as config

#initialize a dispatcher object; dispatcher is strictly speaking
#not necessary, but it simplifies the initialization
dispatcher = Dispatcher(config)    

#initialize data handlers
pip, par, dta, pre, cla = dispatcher.dispatch((Pipeline, FastqParser, DataAnalysisTools, DataPreprocessor, Classifier))
```

First, NGS files need to be parsed to retrive peptide lists. For this, raw R1 and R2 .fastq.gz files have to be merged with FLASH (not covered here;<br>`--max-overlap=200 --allow-outies --max-mismatch-density=0.02`). 

Then, we need to build a FastqParser pipeline to run in silico translation and filtration. Refer to https://github.com/avngrdv/clibas for documentation.

The files are big and may not fit in memory depending on the machine. In such cases, I usually split .gz files in smaller chunks and analyze them one by one.

For this demo, I will parse a single file for speed: `j_LLM_se_r5_m.gz`. This is a round 5 no enzyme selection dataset (track 1 selection). All .gz files should be located in `../sequencing_data` or as specified in config

```python
#enqueue a list of ops to run; at this stage, no data processing will
#take place, but the validity of the specifications will be asserted.

pip.enque([
             par.fetch_gz_from_dir(),
             par.trim_reads(left='ATG', right='CGGAAA'),
             par.translate(force_at_frame=0, stop_readthrough=False),
             dta.length_analysis(where='pep', save_txt=True),
             par.len_filter(where='pep'),
             par.cr_filter(where='pep', loc=[0, 2], tol=5),
             par.vr_filter(where='pep', loc=[1], sets=[1, 2]),
             par.q_score_filt(minQ=30, loc=[1]),
             par.fetch_at(where='pep', loc=[1]),
             par.filt_ambiguous(where='pep'),
             dta.sequence_convergence_analysis(where='pep'),
             par.library_design_match(where='pep'),
             par.unpad(),
             par.save(where='pep', fmt='npy')
         ])

#this will execute the pipeline
#if save_summary=True, summary will be saved in 
#the logs folder as specified in the config file
data = pip.run(save_summary=True)
```

The final track 1 dataset contains `9030589` sequences, before duplicates are removed. A folder that holds parser results is created in `../parser_outputs`. The peptide list is stored in `../parser_outputs/j_LLM_se_r5_m/j_LLM_se_r5_m_pep.npy`. I will move it into `../npy files`. I will also copy the corresponding antiselection peptide list (track 2 data) to `../npy files/j_LLM_an_r5_m_pep.npy`; both will be needed to build a denoiser.

Below the files are preprocessed for model training. The resulting files are quite big: 195 Gb+

```python
#load and preprocess track 1 and 2 datasets:
from clibas.datatypes import Data, AnalysisSample

pos = np.load(os.path.join('../npy files', 'j_LLM_se_r5_m_pep.npy'))
neg = np.load(os.path.join('../npy files', 'j_LLM_an_r5_m_pep.npy'))

data = Data([
              AnalysisSample(X=pos, 
                             y=np.ones(pos.shape[0]), 
                             name='j_LLM_sele_r5'
                            ),
    
              AnalysisSample(X=neg, 
                             y=np.zeros(neg.shape[0]),
                             name='j_LLM_anti_r5'
                            ),
            ]
)

#preprocessing pipeline:
#remove duplicates from each list,
#remove the sequences present in both lists
#shuffle, left/right pad to 18-mers
#split into train and test datasets
#and finally represent as ECFP matrices
#to be saved as .hdf5 files
pip.enque([
             pre.int_repr(alphabet=pre.constants.aas),
             pre.pop_intraset(),
             pre.pop_interset(),
             pre.merge(),
             pre.shuffle(),
             pre.x_expand_and_shift(new_x_dim=18),
             pre.tt_split(test_fraction=0.02),
             pre.to_h5(reshape=True, repad=False, chunks=50, return_data=False),         
           ])

pip.run(data)
```

The train/test files are stored in `../ml_data` or as specified in config.
The files can be directly used by the `Classifier` object for training

Here I will run just a single epoch of the denoiser training. 
`../tf_trained_models/trained_denoiser_CNN.h5` holds the weights for the model used in the manuscript.

```python
cla.epochs = 1

pip.enque([cla.fit()])
pip.run()
```

Now this denoiser can be used to filter track 3 and 4 data. I am using "small scale" MaeJ<sub>C</sub> datasets for this demo so that the filtration process is reasonably quick. The peptide lists are stored in `../npy files/Mae_se_r5_pep_demo.npy` and `../npy files/Mae_an_r5_pep_demo.npy`

```python
mae_pos = np.load(os.path.join('../npy files', 'Mae_se_r5_pep_demo.npy'))
mae_neg = np.load(os.path.join('../npy files', 'Mae_an_r5_pep_demo.npy'))

data = Data([
              AnalysisSample(X=mae_pos,
                             y=None, 
                             seq=mae_pos,
                             name='MaeJ_sele_r5'
                            ),
    
              AnalysisSample(X=mae_neg, 
                             y=None, 
                             seq=mae_neg,
                             name='MaeJ_anti_r5'
                            )
           ])

#the peptides need to be preprocessed in the same way
#as above for the denoiser training;

#pre.featurize_X will hold "featurized" peptides in memory;
#this works for the demo, but full datasets will have to be
#dumped to .hdf5 with pre.to_h5 as above
pip.enque([
            pre.int_repr(alphabet=pre.constants.aas),
            pre.pop_intraset(),
            pre.pop_interset(),
            pre.x_expand_and_shift(new_x_dim=18),
            pre.featurize_X(reshape=True, repad=False),
            cla.predict()           
         ])                       

data = pip.run(data)

#denoiser predictions:
p_mae_pos = data[0].proba.ravel()
p_mae_neg = data[1].proba.ravel()

print(f'Total track 3 peptides: {data[0].seq.shape[0]}')
print(f'Track 3 peptides with p < 0.5: {(p_mae_pos < 0.5).sum()}')

print(f'Total track 4 peptides: {data[1].seq.shape[0]}')
print(f'Track 4 peptides with p < 0.5: {(p_mae_neg < 0.5).sum()}')

#the filtration process itself is straightforward:
mae_pos = data[0].seq[p_mae_pos < 0.5]
mae_neg = data[1].seq[p_mae_neg < 0.5]
```

Note that track 4 filtration was not performed in the manuscript; there is close to no difference. Track 4 dataset contains very little noise. In contrast, track 3 data is over 50% false positives.

Negative dataset augmentation can be done like this:

```python
#use only the track 3 sequences which contain a single Dha
mae_aug = mae_pos[(mae_pos == 'd').sum(axis=1) == 1]

#sample 15% of such sequences for augmentation
AUG_FR = 0.15
ind = np.random.choice(mae_aug.shape[0], 
                       size=int(mae_aug.shape[0] * AUG_FR),
                       replace=False
)

mae_aug = mae_aug[ind]

#mutate Dha to anything but Dha
i1, i2 = np.where(mae_aug == 'd')

#par.constants.aas[:-1] is all possible amino acids except Dha
replacement = np.random.choice(par.constants.aas[:-1],
                               size=(i1.size), 
                               replace=True
)   

mae_aug[i1, i2] = replacement
```

```python
print(mae_aug[:5])
```

Augmented sequences no longer have Dha residues. They can be labelled `0` and used for training/validation. I remove all sequences too similar to `val.pep.01-24` validation peptides (Hamming distance < 8).

The info about `val.pep.01-24` is stored in `../npy files/h.105_138_pepos_NUM.npy` and is referenced in config.

On my machine, TF2 fails to release the GPU memory after<br> 
`tf.keras.backend.clear_session()`<br>
`tf.compat.v1.reset_default_graph()`

I usually reinitialize the kernel to train the classifier specified below. Here, I will just load the weights for a fully trained MaeJ<sub>C</sub> model. The cell below contains the code to train the MaeJ<sub>C</sub> model but isn't executed here.

```python
#make positive labels
pos_labels = 1 - p_mae_pos[p_mae_pos < 0.5]

data = Data([
               AnalysisSample(X=mae_pos, 
                              y=pos_labels, 
                              seq=mae_pos,
                              name='MaeJ_positive_filtered_r5'
                             ),  
    
               AnalysisSample(X=mae_neg,
                              y=np.zeros(mae_neg.shape[0]),
                              seq=mae_neg,
                              name=f'MaeJ_negative_filtered_r5'
                             ),
    
               AnalysisSample(X=mae_aug, 
                              y=np.zeros(mae_aug.shape[0]),
                              seq=mae_aug,
                              name='MaeJ_augmented_r5'
                             ),
             ])

#we need a new Classifier object:
mae_cla, = dispatcher.dispatch((Classifier))

#only 3 epochs
mae_cla.epochs = 3

#preprocess and train
pip.enque([        
            pre.int_repr(alphabet=pre.constants.aas),
            pre.pop_validation(min_hd=8), #no need to pop intra- and interset: done above
            pre.merge(),
            pre.shuffle(),
            pre.x_expand_and_shift(new_x_dim=18),
            pre.tt_split(test_fraction=0.02),
            pre.to_h5(reshape=True, repad=False, chunks=5),
            mae_cla.fit()
         ])

data = pip.run(data)
```

```python
#load MaeJC model weights:
mae_cla, = dispatcher.dispatch((Classifier))
mae_cla.load_model_weights('trained_MaeJC_model.h5')
```

To predict the modification efficiencies of `val.sub.01-24` peptides:

```python
val = np.load('../npy files/h.105_138_pepos_SEQ.npy')

data = Data([
             AnalysisSample(X=val, 
                            y=None, 
                            seq=val,
                            name='validation_peptides'
                           )
])

pip.enque([        
            pre.int_repr(alphabet=pre.constants.aas),
            pre.x_expand_and_shift(new_x_dim=18),
            pre.featurize_X(reshape=True, repad=False),
            mae_cla.predict()
         ])

data = pip.run(data)
```
